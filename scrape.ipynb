```python
years = range(2000,2016)
player_data = []
new_columns = column_names
new_columns.append(u'year')
player_data.append(new_columns)
for year in years:
    directory = "http://www.pro-football-reference.com/years/{}/draft.htm".format(str(year))
    text = requests.get(directory).text
    soup = BeautifulSoup(text, "lxml")
    trs = soup.findAll('tr')
    for tr in trs[2:-3]:
        data = [c.text for c in tr.findAll('td')]
        data.append(year)
        player_data.append(data)
```
